```python
  def norm_func(i):
    x = (i-i.min())/(i.max()-i.min())
    return (x)
```


# Data Preprocessing

```python
standard_scaler = StandardScaler()
std_df1 = standard_scaler.fit_transform(df1)
std_df1.shape
```

```python
minmax = MinMaxScaler()
norm_df1 = minmax.fit_transform(df1)
norm_df1.shape
```

# K Means Clustering

```python
import warnings
warnings.filterwarnings('ignore')
fig = plt.figure(figsize=(10, 8))
WCSS = []
for i in range(1, 12):
    clf = KMeans(n_clusters=i)
    clf.fit(norm_df1)
    WCSS.append(clf.inertia_) # inertia is another name for WCSS
plt.plot(range(1, 12), WCSS)
plt.title('The Elbow Method')
plt.ylabel('WCSS')
plt.xlabel('Number of Clusters')
plt.show()  
```

```python
WCSS
```

```python
clf = KMeans(n_clusters=4)
y_kmeans = clf.fit_predict(norm_df1)  
y_kmeans
```

# HIERARCHAICAL Clustering

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df_ss = pd.DataFrame(scaler.fit_transform(df1.iloc[:,1:7]))
df_ss
```

```python
from scipy.cluster.hierarchy import linkage 
import scipy.cluster.hierarchy as sch # for creating dendrogram 
p = np.array(df1_norm) # converting into numpy array format 
z = linkage(df1_norm, method="single",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
    #leaf_rotation=6.,  # rotates the x axis labels
    #leaf_font_size=15.,  # font size for the x axis labels
)
plt.show()  
```

```python
p = np.array(df1_norm) 
z = linkage(df1_norm, method="complete",metric="euclidean")
plt.figure(figsize=(15, 5))
plt.title('Hierarchical Clustering Dendrogram')
plt.xlabel('Index')
plt.ylabel('Distance')
sch.dendrogram(
    z,
)
plt.show()    
```
