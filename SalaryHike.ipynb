```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as smf
```

```python
df= pd.read_csv("C:\\Users\\pradyumn\\Documents\\Assignment\\Salary_Data.csv")
df.head()
```

```python
X=df[["YearsExperience"]]
Y=df[["Salary"]]
```

```python
plt.scatter(X,Y)
plt.ylabel("Salary")
plt.xlabel("YearsExperience")
plt.show()
```

```python
df.hist()
```

```python
df.boxplot()
```

```python
df.corr()
```

```python
sns.regplot(x=X,y=Y)
```

```python
model=smf.ols("Y~X",data=df).fit()
model.summary()
```

# Model Testing


```python
model.params
```

```python
model.tvalues , model.pvalues
```

```python
model.rsquared , model.rsquared_adj
```

```python
import statsmodels.formula.api as smf
model2 = smf.ols("np.log(Y)~X",data=df).fit() 
model2.params
model2.summary()
```

```python
(model2.rsquared,model2.rsquared_adj)
```

```python
rmse_residuals = np.sqrt(model2.mse_resid)
rmse_residuals
```

```python
model3 = smf.ols("np.sqrt(Y)~X", data=df).fit() 
model3.params
model3.summary()     
```

```python
(model3.rsquared,model3.rsquared_adj)
```

```python
rmse_residuals = np.sqrt(model3.mse_resid)
rmse_residuals
```

# Hence  Rsquared value is not improving by using transformation technique so and Rsquared isbetter in simple model
