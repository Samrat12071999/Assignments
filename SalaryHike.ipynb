```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import statsmodels.formula.api as smf
```

```python
df= pd.read_csv("C:\\Users\\pradyumn\\Documents\\Assignment\\Salary_Data.csv")
df.head()
```

```python
X=df[["YearsExperience"]]
Y=df[["Salary"]]
```

```python
plt.scatter(X,Y)
plt.ylabel("Salary")
plt.xlabel("YearsExperience")
plt.show()
```

```python
df.hist()
```

```python
df.boxplot()
```

```python
df.corr()
```

```python
sns.regplot(x=X,y=Y)
```

```python
model=smf.ols("Y~X",data=df).fit()
model.summary()
```
